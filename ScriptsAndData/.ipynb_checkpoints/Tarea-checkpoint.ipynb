```python
#importación de las librerias que usaremos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#cargamos el dataset
data=pd.read_csv("data/dengue_clima.csv") 
```

```python
data.head()
```

```python
data.describe()
```

```python
#Seleccionamos las variables que debemos usar para trabajar
# seleccionar variables de la solicitud, sistema financiero y target, descartar variables pos aprobación
df1 = data.iloc[:,11:22]
df2 = data.iloc[:,33:77]
df3 = data.iloc[:,88:100]

# Unificamos en un dataframe filtrado
data2 = pd.concat([df1,df2,df3], axis=1)
data2.head()
```

```python
data2.describe()
```

```python
# Matriz de covarianzas
data2.cov()

# Calcular variables con correlacion positiva o negativa superior a un umbral
corMatrix = data2.corr()
corMatrix[(abs(corMatrix) > 0.30) & (corMatrix != 1)]['cantidad'].dropna()
```

```python
# Scatterplot
data2.plot.scatter(x='temperatura_media_media(-1)', y='cantidad', c='DarkBlue')
data2.plot.scatter(x='humedad_min_media_porc(-1)', y='cantidad', c='DarkBlue')
data2.plot.scatter(x='humedad_media_media_porc(-1)', y='cantidad', c='DarkBlue')
data2.plot.scatter(x='humedad_max_media_porc(-1)', y='cantidad', c='DarkBlue')
data2.plot.scatter(x='lluvia_mm(-1)', y='cantidad', c='DarkBlue')
data2.plot.scatter(x='nivel(-1)', y='cantidad', c='DarkBlue')
```

```python
# Cargar funciones de la librería de python sklearn y math
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt

# Seleccionar variables correlacionadas
df = data2[['temperatura_media_media(-7)','temperatura_media_media(-8)','temperatura_media_media(-9)','temperatura_media_media(-10)','temperatura_media_media(-11)','nivel(-5)','nivel(-6)','nivel(-7)','nivel(-8)','nivel(-9)','nivel(-10)','nivel(-11)','cantidad']]

# Seleccionar x e y
x = pd.DataFrame(df[['temperatura_media_media(-7)','temperatura_media_media(-8)','temperatura_media_media(-9)','temperatura_media_media(-10)','temperatura_media_media(-11)','nivel(-5)','nivel(-6)','nivel(-7)','nivel(-8)','nivel(-9)','nivel(-10)','nivel(-11)']]).iloc[0:262]
y = pd.DataFrame(df['cantidad']).iloc[0:262]
x_val = pd.DataFrame(df[['temperatura_media_media(-7)','temperatura_media_media(-8)','temperatura_media_media(-9)','temperatura_media_media(-10)','temperatura_media_media(-11)','nivel(-5)','nivel(-6)','nivel(-7)','nivel(-8)','nivel(-9)','nivel(-10)','nivel(-11)']]).iloc[263:]
y_val = pd.DataFrame(df['cantidad']).iloc[263:]

# entrenar el modelo y calcular predicciones
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x_val)

# graficar variables y predicción lineal
#plt.scatter(x_val, y_val)
#plt.plot(x_val, y_pred, color='red')
#plt.show()

# calcular coeficiente r2 y pearson
print("Coeficiente r2: " + str(model.score(x, y)))
print("Coeficiente de Pearson (r): " + str(sqrt(model.score(x, y))))

# calcular root mean square error (RMSE)
print("Raíz del Error cuadrático Medio (RMSE): " + str(sqrt(mean_squared_error(y_val, y_pred))))
print("Media de Nasdaq: " + str(y_val['cantidad'].mean()))
print("% de error: " + str(sqrt(mean_squared_error(y_val, y_pred)) / y_val['cantidad'].mean()))
```

## Conclusiones
Este caso no puede ser representado por una regresion lineal pues los factores que intervienen en el Dengue no puede ser representados mediante lineas rectas

## Método PCA

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
import pandas as pd  

x = data.iloc[0:200,1:66]
x_val = data.iloc[200:,1:66]
y = data.iloc[0:200,66:]
y_val = data.iloc[200:,66:]

# crear PCA que explica el 98% de la varianza acumulada 
pca = PCA(n_components = 13)
x = pca.fit_transform(x)

# Imprimir componentes y varianza acumulada
print(x.shape)
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.cumsum())
#print(pca.components_)

# Transformar x_val
x_val = pca.transform(x_val)

# entrenar el modelo y calcular predicciones
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x_val)

# calcular coeficiente r2 y pearson
print("Coeficiente r2: " + str(model.score(x, y)))
print("Coeficiente de Pearson (r): " + str(sqrt(model.score(x, y))))

# calcular root mean square error (RMSE)
print("Raíz del Error cuadrático Medio (RMSE): " + str(sqrt(mean_squared_error(y_val, y_pred))))
print("Media de Nasdaq: " + str(y_val['cantidad'].mean()))
print("% de error: " + str(sqrt(mean_squared_error(y_val, y_pred)) / y_val['cantidad'].mean()))
```

## Conlusiones
Mediante este modelo podemos conseguir un mejor resultado que mediante la regresión lineal, para ello hemos identificado que debemos usar 13 variables a fin de obtener un 98% de acierto
